```python
import matplotlib.pyplot as plt
import pandas as pd
import os
models = ['exel/8-640', 'exel/8-640 - yolo9t', 'exel/8-640 - yolo10n', 'exel/8-640 - yolo11n']
labels = ['YOLOv8n', 'YOLO9n', 'YOLO10n', 'YOLO11n']
colors = ['blue', 'orange', 'green', 'red']
linestyle=['--', '-', '-.', '--']

plt.figure(figsize=(8, 8))

for i, model_path in enumerate(models):
    csv_path = os.path.join(model_path, 'results.csv')
        
    df = pd.read_csv(csv_path)
        
    epochs = df['epoch'] 
    accuracy = df['metrics/mAP50-95(B)']
    mask = epochs <= 10
    plt.plot(epochs[mask], accuracy[mask], label=labels[i], color=colors[i], linewidth=2, linestyle=linestyle[i])
  
plt.title('Model Accuracy Comparison', fontsize=16)
plt.xlabel('Epochs', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()

```
