```python
import matplotlib.pyplot as plt
import pandas as pd
import os


models = ['exel/8-640 - lr0-0.005', 'exel/8-640 - lr0-0.01', 'exel/8-640 - lr0-0.001', 'exel/8-640 - lr0-0.1']
labels = ['LR=0.005', 'LR=0.01', 'LR=0.001', 'LR=0.1']
colors = ['blue', 'yellow', 'green', 'red']
linestyle=['--', '-', '-.', '--']

plt.figure(figsize=(10, 10))

for i, model_path in enumerate(models):
    csv_path = os.path.join(model_path, 'results.csv')
    
    df = pd.read_csv(csv_path)
        
    epochs = df['epoch'] 
    accuracy = df['metrics/mAP50-95(B)']
        
     
    plt.plot(epochs, accuracy,color=colors[i], label=labels[i], linewidth=2, linestyle=linestyle[i])
plt.title('Model Accuracy Comparison', fontsize=16)
plt.xlabel('Epochs', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd
import os


models = ['exel/8-640 - lr0-0.005',  'exel/8-640 - lr0-0.001', 'exel/8-640 - lr0-0.1']
labels = ['LR=0.005',  'LR=0.001', 'LR=0.1']
colors = ['blue',  'green', 'red']
linestyle=['--', '-', '-.', '--']

plt.figure(figsize=(10, 10))

for i, model_path in enumerate(models):
    csv_path = os.path.join(model_path, 'results.csv')
    df = pd.read_csv(csv_path)
        
    epochs = df['epoch'] 
    accuracy = df['metrics/mAP50-95(B)'] 
  
    plt.plot(epochs, accuracy,color=colors[i], label=labels[i], linewidth=2, linestyle=linestyle[i])
plt.title('Model Accuracy Comparison', fontsize=16)
plt.xlabel('Epochs', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()

```
